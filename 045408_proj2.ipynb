```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
label_encoder = LabelEncoder()
df['order_status'] = label_encoder.fit_transform(df['order_status'])
df['payment_type'] = label_encoder.fit_transform(df['payment_type'])

# Handle missing values
df.fillna(df.mean(), inplace=True)
df
```

```python
df.info()
```

```python
#subsetting
df_inputs = df[['order_status', 'payment_sequential', 'payment_type',
       'payment_installments', 'price', 'freight_value', 'Age', 'review_score']]
print(df_inputs)
df_output = df[['cluster']]; df_output
```

```python
df_inputs_names = df_inputs.columns
print(df_inputs_names)

df_output_labels = df_output['cluster'].unique().astype(str)
df_output_labels

```

```python
# Split the mtcars Data Subset into Training & Testing Sets
train_df_inputs, test_df_inputs, train_df_output, test_df_output = train_test_split(df_inputs, df_output, test_size=0.25, random_state=45048)
```

```python
# Initialize StratifiedShuffleSplit with desired test size and random state
stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=45048)

# Perform the stratified split to get training and testing indices
for train_index, test_index in stratified_split.split(df_inputs, df_output):
    df_inputs_train, df_inputs_test = df_inputs.iloc[train_index], df_inputs.iloc[test_index]
    df_output_train, df_output_test = df_output.iloc[train_index], df_output.iloc[test_index]
```

```python
# Decision Tree : Model (Training Subset)
dtc = DecisionTreeClassifier(criterion='gini', random_state=45048) # Other Criteria : Entropy,  Log Loss
dtc_model = dtc.fit(train_df_inputs, train_df_output)
dtc_model
```

```python
# Decision Tree : Model Rules
dtc_model_rules = export_text(dtc_model, feature_names = list(df_inputs_names)); print(dtc_model_rules)
```

```python
# Decision Tree : Feature Importance
dtc_imp_features = pd.DataFrame({'feature': df_inputs_names, 'importance': np.round(dtc_model.feature_importances_, 3)})
dtc_imp_features.sort_values('importance', ascending=False, inplace=True)
dtc_imp_features

```

```python
# Decision Tree : Model Prediction (Training Subset)
dtc_model_predict = dtc_model.predict(train_df_inputs)
dtc_model_predict

# Memory usage
memory_used = memory_usage()

# Print memory usage and execution time
print("Memory used (MB):", memory_used)
```

```python
# Decision Tree : Prediction (Testing Subset)
dtc_predict = dtc_model.predict(test_df_inputs)
dtc_predict

# Memory usage
memory_used = memory_usage()

# Print memory usage and execution time
print("Memory used (MB):", memory_used)
```

```python
# Decision Tree : Model Evaluation (Training Subset)
dtc_model_conf_mat = pd.DataFrame(confusion_matrix(train_df_output, dtc_model_predict)); print(dtc_model_conf_mat)
dtc_model_perf = classification_report(train_df_output, dtc_model_predict); print(dtc_model_perf)

# Memory usage
memory_used = memory_usage()

# Print memory usage and execution time
print("Memory used (MB):", memory_used)
```

```python
 #Decision Tree : Prediction Evaluation (Testing Subset)
dtc_predict_conf_mat = pd.DataFrame(confusion_matrix(test_df_output, dtc_predict)); print(dtc_predict_conf_mat)
dtc_predict_perf = classification_report(test_df_output, dtc_predict); print(dtc_predict_perf)

```

```python
# Define the values of k
k_values = [7,9,11]

for k in k_values:
    # Create the KNN Model
    knn = KNeighborsClassifier(n_neighbors=k)

    # Train the Model
    knn.fit(train_df_inputs, train_df_output)

    # Make Predictions
    knn_predict_test = knn.predict(test_df_inputs)

    # Evaluate the Model
    knn_conf_mat_test = confusion_matrix(test_df_output, knn_predict_test)
    knn_perf_test = classification_report(test_df_output, knn_predict_test)
    print(f"KNN Model Performance on Testing Set (k={k}):\n", knn_perf_test)

    # Visualize Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(knn_conf_mat_test, annot=True,fmt='d', cmap='Paired')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'KNN: Confusion Matrix (Testing Set, k={k})')
    plt.show()

    # Compute Accuracy
    knn_accuracy = accuracy_score(test_df_output, knn_predict_test)
    print(f"KNN Model Accuracy (k={k}):", knn_accuracy)
    print('\n')
```
